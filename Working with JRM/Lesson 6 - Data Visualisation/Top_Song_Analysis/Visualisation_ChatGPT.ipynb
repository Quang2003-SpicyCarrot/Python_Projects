```python
import pandas as pd
import os
#import pip
#pip.main(['install','seaborn'])
import seaborn as sb
import matplotlib.pyplot as plt
import pathlib
```

```python
# Directory containing the CSV files
__file__ = 'modulename.py'
directory = pathlib.Path(__file__).parent

# List to store DataFrames
dataframes_list = []

# Iterate over each file in the directory
for filename in os.listdir(directory):
    if filename.endswith('.csv'):
        # Construct full file path
        file_path = os.path.join(directory, filename)
        # Read the CSV file and append to list
        dataframes_list.append(pd.read_csv(file_path))

combined_dataframe = pd.concat(dataframes_list)
combined_dataframe = combined_dataframe.rename(columns = {"Unnamed: 0": "id"})
combined_dataframe = combined_dataframe.drop_duplicates(subset='id')
combined_dataframe
```

```python
# Directory containing the CSV files
__file__ = 'modulename.py'
directory = pathlib.Path(__file__).parent

# List to store DataFrames
dataframes_list = []

# Iterate over each file in the directory
for filename in os.listdir(directory):
    if filename.endswith('.csv'):
        # Construct full file path
        file_path = os.path.join(directory, filename)
        # Read the CSV file and append to list
        dataframes_list.append(pd.read_csv(file_path))

# Concatenate all the DataFrames in the list
top_spotify_songs = pd.concat(dataframes_list)
top_spotify_songs = top_spotify_songs.rename(columns = {"Unnamed: 0": "id"})
top_spotify_songs = top_spotify_songs.drop_duplicates(subset='id')
top_spotify_songs
```

```python
# Code with ChatGPT
import matplotlib.pyplot as plt
import seaborn as sns

# Setting the aesthetic style of the plots
sns.set(style="whitegrid")

# Creating the visualizations

# 1. Line Chart: Trend of Popularity Over Years
plt.figure(figsize=(5, 3))
sns.lineplot(data=top_spotify_songs, x='year', y='pop', ci=None)
plt.title('Trend of Popularity Over Years')
plt.xlabel('Year')
plt.ylabel('Popularity')
plt.show()

# 2. Bar Chart: Distribution of Top Genres
plt.figure(figsize=(5, 3))
top_genres = top_spotify_songs['top genre'].value_counts().head(10)
sns.barplot(x=top_genres.values, y=top_genres.index)
plt.title('Top 10 Genres in Top Spotify Songs')
plt.xlabel('Number of Songs')
plt.ylabel('Genre')
plt.show()

# 3. Scatter Plot: Energy vs. Danceability
plt.figure(figsize=(5, 3))
sns.scatterplot(data=top_spotify_songs, x='nrgy', y='dnce')
plt.title('Energy vs. Danceability of Top Spotify Songs')
plt.xlabel('Energy')
plt.ylabel('Danceability')
plt.show()

# 4. Histogram: Distribution of Song Durations
plt.figure(figsize=(5, 3))
sns.histplot(top_spotify_songs['dur'], kde=True)
plt.title('Distribution of Song Durations')
plt.xlabel('Duration (Seconds)')
plt.ylabel('Frequency')
plt.show()

# 5. Box Plot: Loudness Across Years
plt.figure(figsize=(5, 3))
sns.boxplot(data=top_spotify_songs, x='year', y='dB')
plt.title('Loudness (dB) Across Years')
plt.xlabel('Year')
plt.ylabel('Loudness (dB)')
plt.show()
```

```python
# Preparing data for the additional visualizations

# 6. Data for Pie Chart: Share of Top 5 Genres
top_5_genres = combined_dataframe['top genre'].value_counts().head(5)

# 7. Data for Heatmap: Correlation of Song Attributes
correlation_data = combined_dataframe[['nrgy', 'dnce', 'dB', 'live', 'val', 'dur', 'acous', 'spch', 'pop']].corr()

# 8. Data for Bubble Chart: Popularity, Energy, and Danceability
# Using a sample for clearer visualization
bubble_chart_data = combined_dataframe.sample(n=100)

# 9. & 10. Data for Area Charts: Energy and Genres Over Years
# Short Data
combined_dataframe_short = combined_dataframe.drop(columns=['title', 'artist', 'top genre'])
# Aggregating data by year
yearly_data = combined_dataframe_short.groupby('year').mean()
# Preparing data for stacked area chart
top_genres_yearly = combined_dataframe[combined_dataframe['top genre'].isin(top_5_genres.index)]
stacked_area_data = top_genres_yearly.pivot_table(index='year', columns='top genre', values='id', aggfunc='count', fill_value=0)

# Creating the visualizations

# 6. Pie Chart: Share of Top 5 Genres
plt.figure(figsize=(7, 7))
plt.pie(top_5_genres, labels=top_5_genres.index, autopct='%1.1f%%', startangle=140)
plt.title('Share of Top 5 Genres in Top Spotify Songs')
plt.show()

# 7. Heatmap: Correlation of Song Attributes
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_data, annot=True, cmap='coolwarm')
plt.title('Correlation of Song Attributes')
plt.show()

# 8. Bubble Chart: Popularity, Energy, and Danceability
plt.figure(figsize=(10, 6))
plt.scatter('nrgy', 'dnce', s='pop', alpha=0.5, data=bubble_chart_data)
plt.xlabel('Energy')
plt.ylabel('Danceability')
plt.title('Popularity, Energy, and Danceability of Top Spotify Songs')
plt.show()

# 9. Area Chart: Energy Over Years
plt.figure(figsize=(10, 6))
plt.fill_between(yearly_data.index, yearly_data['nrgy'], color="skyblue", alpha=0.4)
plt.plot(yearly_data.index, yearly_data['nrgy'], color="Slateblue", alpha=0.6, linewidth=2)
plt.title('Average Energy of Songs Over Years')
plt.xlabel('Year')
plt.ylabel('Average Energy')
plt.show()

# 10. Stacked Area Chart: Distribution of Top 5 Genres Over Years
plt.figure(figsize=(10, 6))
plt.stackplot(stacked_area_data.index, stacked_area_data.T, labels=stacked_area_data.columns)
plt.title('Distribution of Top 5 Genres Over Years')
plt.xlabel('Year')
plt.ylabel('Number of Songs')
plt.legend(loc='upper left')
plt.show()
```
