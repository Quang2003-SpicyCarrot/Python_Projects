# Import necessary libraries

```python
import pandas as pd
import os
#import pip
#pip.main(['install','seaborn'])
import seaborn as sb
import matplotlib.pyplot as plt
import pathlib
```

# Read and save all .csv file

From the local repository, we import all the .csv file stored in there and name the created variable with "top_spotify_songs".

```python
# Directory containing the CSV files
__file__ = 'modulename.py'
directory = pathlib.Path(__file__).parent

# List to store DataFrames
dataframes_list = []

# Iterate over each file in the directory
for filename in os.listdir(directory):
    if filename.endswith('.csv'):
        # Construct full file path
        file_path = os.path.join(directory, filename)
        # Read the CSV file and append to list
        dataframes_list.append(pd.read_csv(file_path))

# Concatenate all the DataFrames in the list
top_spotify_songs = pd.concat(dataframes_list)
top_spotify_songs = top_spotify_songs.rename(columns = {"Unnamed: 0": "id"})
top_spotify_songs = top_spotify_songs.drop_duplicates(subset='id')
top_spotify_songs.head(5)
```

#### 2. Check the data type

```python
top_spotify_songs.dtypes
```

#### 3. What is the most popular song in the dataset?

As to check the most popular song in the dataset, we need to check the max number of counts after sorting data by "ID" to avoid 2 or more songs have the same values. Then, by displaying the max count, we can find out the most popular song.

```python
top_spotify_songs.sort_values(['pop'], ascending = False).head(1)
```

Hence, the most popular song is "Memories" by Maroon 5

#### 4. What artist has the most songs in the dataset?

To check which artist shares the most songs in the dataset, we count the ids (which is the unique ids of each song) of each artist. Then we sort values according to the max count of ids of each artist and display the highest count.

```python
top_spotify_songs.groupby(['artist'])['id'].count().sort_values(ascending = False).reset_index().head(1)
```

Hence the artist who has the most songs is Katty Perry

#### 5. What is the most popular genre of music in the dataset?

As similar to the method shared in the previous part. We count the ids of each genre

```python
top_spotify_songs.groupby(['top genre'])['id'].count().sort_values(ascending = False).reset_index().head(1)
```

Therefore, the most popular genre of music in the dataset is dance pop with the total id counts of 327.

####  6. Are there any songs in the dataset more than once?

```python
top_spotify_songs.groupby(['title'])['id'].count().sort_values(ascending = False).reset_index().head(1)
```

Based on the given dataset, the song that appeared twice is "I like it".

#### 7. Is there a relationship between popularity and danceability?
To answear this question, I create a scatter plot to see the relationship between popularity values and danceability values. You can see this relationship is not strong between popularity values and danceability values  

```python
sb.regplot(data = top_spotify_songs, 
           x='dnce', 
           y='pop', 
           color="pink")

plt.title('Relationship between popularity and danceability', fontsize = 15)
plt.xlabel("Danceability")
plt.ylabel("Popularity")
plt.show()
```

There is a cohesive random cluster around 40 to 100 in Danceability and between 25 and 100 in Popularity. This suggests that the popularity of a song has a correlation with its danceability, it also means that many people prefer songs with higher danceability.

#### 8. Does having a higher bpm lead to having a higher energy score?
To answear this question, I also using the scatter plot to show the relationship between bpm values and enerfy score values.
<br> As you can see, this is hardly to tell that bpm values and energy score have relationship
<br> because the regression nearly resembles the x-axis but with a constant of 100

```python
sb.regplot(data = top_spotify_songs, 
           x='nrgy', 
           y='bpm', 
           color="pink")

plt.title('Relationship between bpm and energy score', fontsize = 15)
plt.xlabel("Energy")
plt.ylabel("BPM")
plt.show()
```

According to the generated graph which described the relationship between BPM and Energy, a correlated pattern between the energy score and BPM is discovered. The energy scores better as the beat per minute of songs ranges in between around 75 to approximately 200 BPM. based on the given dataset, there are more songs with BPM ranging between 75 to around 150 and the energy gradually increases from at least 40 to 100 at peak point.

It can also be said that the high energy score clusters around 40 to 100 where BPM is in-between 7 to approximately 200.

#### 9. Which genre has the highest energy score?
To see which genre has the highest energy score, we check for the average energy score for each song and then we sort out the top song which produces the highest energy score.

```python
top_spotify_songs.groupby(['top genre'])['nrgy'].mean().sort_values(ascending = False).head(1).reset_index()
```

```python
highest_nrgy_genre = top_spotify_songs.groupby(['top genre'])['nrgy'].mean().sort_values(ascending = False).head(10).reset_index()

plt.figure(figsize = (10, 5))
ax = sb.barplot(data = highest_nrgy_genre,
           x = "top genre",
           y = "nrgy")

for i in ax.containers:
    ax.bar_label(i, size = 10)

plt.xlabel("top genre", size = 14)
plt.xticks(rotation = 60)
plt.ylabel("energy", size = 14)
plt.title("Top genres with the highest average energy score", size = 18)
```

Based on the top 10 genres, french indie pop has the highest average energy score of 89, and boy band is at the lowest with the score of 80.26.

#### 10. What percentage of songs may have been recorded live? (Consider anything above 50 in the live column to be recorded live)

#### 11. What percentage of songs were acoustic? (Consider anything above 50 in the acoustic column to be acoustic)

#### 12. Which years songs have the highest amount of spoken word in the song?

```python
top_spotify_songs.groupby(['year'])['spch'].sum().sort_values(ascending = False).reset_index().head(1)
```

```python
# Code with ChatGPT
import matplotlib.pyplot as plt
import seaborn as sns

# Setting the aesthetic style of the plots
sns.set(style="whitegrid")

# Creating the visualizations

# 1. Line Chart: Trend of Popularity Over Years
plt.figure(figsize=(5, 3))
sns.lineplot(data=top_spotify_songs, x='year', y='pop', ci=None)
plt.title('Trend of Popularity Over Years')
plt.xlabel('Year')
plt.ylabel('Popularity')
plt.show()

# 2. Bar Chart: Distribution of Top Genres
plt.figure(figsize=(5, 3))
top_genres = top_spotify_songs['top genre'].value_counts().head(10)
sns.barplot(x=top_genres.values, y=top_genres.index)
plt.title('Top 10 Genres in Top Spotify Songs')
plt.xlabel('Number of Songs')
plt.ylabel('Genre')
plt.show()

# 3. Scatter Plot: Energy vs. Danceability
plt.figure(figsize=(5, 3))
sns.scatterplot(data=top_spotify_songs, x='nrgy', y='dnce')
plt.title('Energy vs. Danceability of Top Spotify Songs')
plt.xlabel('Energy')
plt.ylabel('Danceability')
plt.show()

# 4. Histogram: Distribution of Song Durations
plt.figure(figsize=(5, 3))
sns.histplot(top_spotify_songs['dur'], kde=True)
plt.title('Distribution of Song Durations')
plt.xlabel('Duration (Seconds)')
plt.ylabel('Frequency')
plt.show()

# 5. Box Plot: Loudness Across Years
plt.figure(figsize=(5, 3))
sns.boxplot(data=top_spotify_songs, x='year', y='dB')
plt.title('Loudness (dB) Across Years')
plt.xlabel('Year')
plt.ylabel('Loudness (dB)')
plt.show()
```

```python
# Preparing data for the additional visualizations

# 6. Data for Pie Chart: Share of Top 5 Genres
top_5_genres = top_spotify_songs['top genre'].value_counts().head(5)

# 7. Data for Heatmap: Correlation of Song Attributes
correlation_data = top_spotify_songs[['nrgy', 'dnce', 'dB', 'live', 'val', 'dur', 'acous', 'spch', 'pop']].corr()

# 8. Data for Bubble Chart: Popularity, Energy, and Danceability
# Using a sample for clearer visualization
bubble_chart_data = top_spotify_songs.sample(n=100)

# 9. & 10. Data for Area Charts: Energy and Genres Over Years
# Short Data
combined_dataframe_short = top_spotify_songs.drop(columns=['title', 'artist', 'top genre'])
# Aggregating data by year
yearly_data = combined_dataframe_short.groupby('year').mean()
# Preparing data for stacked area chart
top_genres_yearly = top_spotify_songs[top_spotify_songs['top genre'].isin(top_5_genres.index)]
stacked_area_data = top_genres_yearly.pivot_table(index='year', columns='top genre', values='id', aggfunc='count', fill_value=0)

# Creating the visualizations

# 6. Pie Chart: Share of Top 5 Genres
plt.figure(figsize=(7, 7))
plt.pie(top_5_genres, labels=top_5_genres.index, autopct='%1.1f%%', startangle=140)
plt.title('Share of Top 5 Genres in Top Spotify Songs')
plt.show()

# 7. Heatmap: Correlation of Song Attributes
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_data, annot=True, cmap='coolwarm')
plt.title('Correlation of Song Attributes')
plt.show()

# 8. Bubble Chart: Popularity, Energy, and Danceability
plt.figure(figsize=(10, 6))
plt.scatter('nrgy', 'dnce', s='pop', alpha=0.5, data=bubble_chart_data)
plt.xlabel('Energy')
plt.ylabel('Danceability')
plt.title('Popularity, Energy, and Danceability of Top Spotify Songs')
plt.show()

# 9. Area Chart: Energy Over Years
plt.figure(figsize=(10, 6))
plt.fill_between(yearly_data.index, yearly_data['nrgy'], color="skyblue", alpha=0.4)
plt.plot(yearly_data.index, yearly_data['nrgy'], color="Slateblue", alpha=0.6, linewidth=2)
plt.title('Average Energy of Songs Over Years')
plt.xlabel('Year')
plt.ylabel('Average Energy')
plt.show()

# 10. Stacked Area Chart: Distribution of Top 5 Genres Over Years
plt.figure(figsize=(10, 6))
plt.stackplot(stacked_area_data.index, stacked_area_data.T, labels=stacked_area_data.columns)
plt.title('Distribution of Top 5 Genres Over Years')
plt.xlabel('Year')
plt.ylabel('Number of Songs')
plt.legend(loc='upper left')
plt.show()
```
