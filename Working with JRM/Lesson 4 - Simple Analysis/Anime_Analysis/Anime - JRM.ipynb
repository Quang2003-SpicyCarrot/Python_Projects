# Dữ liệu
Dữ liệu cho bài thực hành này được tổng hợp từ [MyAnimeList](https://myanimelist.net/) được lưu trong file anime_info.csv và bao gồm các cột:
* `anime_id`: mã số của anime trên [MyAnimeList](https://myanimelist.net/).
* `title` : tựa của anime.
* `type` : hình thức phát hành (TV Series, Movie, OVA, ...).
* `source` : nguyên tác của anime (Manga, Light Novel, Game, ...).
* `episodes` : số tập.
* `duration_min` : thời lượng mỗi tập (tính bằng phút).
* `score` : điểm của anime trên [MyAnimeList](https://myanimelist.net/), tổng hợp từ đánh giá có người dùng.
* `studio` : hãng sản xuất.
* `genre` : thể loại phim.
* `start` : ngày bắt đầu phát hành.
* `finish` : ngày kết thúc phát hành.

# Bài tập

**Câu 1**: Đọc dữ liệu từ file csv và lưu dưới tên `anime_data`.

```python
import pandas as pd
import numpy as np
import datetime
anime_data = pd.read_csv('Anime_Data.txt')
anime_data
```

**Câu 2**: Thực hiện việc chuyển đổi cần thiết.

```python
anime_data = anime_data.assign(
    score = pd.to_numeric(anime_data.score),
    start = pd.to_datetime(anime_data.start),
    finish = pd.to_datetime(anime_data.finish)
)
anime_data.dtypes
```

**Câu 3**: Xóa những dòng có ngày bắt đầu hoặc kết thúc là `NaT`.

```python
anime_data = anime_data.dropna(axis = 0,how ='any')
anime_data
```

```python
anime_data = anime_data[anime_data['duration_min'] != 0]
anime_data
```

**Câu 9**: Thêm vào dữ liệu cột `season` để chỉ mùa phát sóng (tính theo ngày bắt đầu phát hành) của anime với:

|Ngày phát hành|Season|
|---|---|
|Từ 01.01 đến 31.03|Winter-Spring|
|Từ 01.04 đến 30.06|Spring|
|Từ 01.07 đến 30.09|Summer|
|Từ 01.10 đến 31.12|Fall|

In ra `season` có nhiều bộ anime phát hành nhất.


```python
def name_season(month):
  if month <= 3:
    return 'Winter-Spring'
  elif 3 < month <= 6:
    return 'Spring'
  elif 6 < month <= 9:
    return 'Summer'
  else:
    return 'Fall'
anime_data = anime_data.assign(
    season = anime_data.start.dt.month.apply(name_season)
)
```

```python
anime_season = []
season = anime_data.season.unique()
for name_season in season:
  data = anime_data[anime_data['season'] == name_season].shape[0]
  anime_season.append(data)
```

```python
max_anime_season = pd.DataFrame()
max_anime_season = max_anime_season.assign(
    season = season,
    number = anime_season
)
max_anime_season.sort_values(by = 'number' ,ascending = False).iloc[0].season
max_anime_season
```

Như có thể thấy, `genre` của những anime là một danh sách nhiều thể loại khác nhau. Điều này gây khó khăn cho việc tổng hợp thông tin theo từng `genre`. Vì vậy, phải phân rã cột `genre` sao cho mỗi dòng chỉ chứa 1 `genre` như minh họa:
![Explode](https://www.w3resource.com/w3r_images/pandas-dataframe-explode-1.png)

```python
# thực hiện đoạn code sau đây để phân rã cột `genre`
# chuyển đổi cột `genre` từ str sang list
movie = movie.assign(genre = movie.genre.str.split(', '))
# phân rã cột genre
movie = movie.explode('genre')
movie
```

**Câu 12**: Thực hiện các yêu cầu sau:
1. Thống kê số lượng `genre`.
2. In ra điểm trung bình và số lượng anime từng `genre`.
3. Thống kê điểm trung bình theo `studio` và `genre`.


```python
movie.groupby(by = 'genre').count().anime_id
```

```python
genre_movie = movie['genre'].unique()
number_by_genre = []
score_by_genre = []
for genre in genre_movie:
    number_by_genre.append(movie[movie['genre'] == genre].genre.count())
    score_by_genre.append(round(movie[movie['genre'] == genre].score.mean(),2))
dict = {'Genre Movie': genre_movie,'Number Movie': number_by_genre,'Genre Score': score_by_genre, }
data_movie_anime = pd.DataFrame(dict)
data_movie_anime
```

```python
studio_movie = movie['studio'].unique()
list_studio_genre = []
score_studio_genre = []
for studio in studio_movie:
    studio_movie = movie[movie['studio'] == studio]
    for genre in genre_movie:
        list_studio_genre.append(studio + ' ' + genre)
        score_studio_genre.append(round(studio_movie[studio_movie['genre'] == genre].score.mean(),2))
dict = {'Studio by Genre': list_studio_genre, 'Score': score_studio_genre}
data = pd.DataFrame(dict)
with pd.option_context('display.max_rows', None,
                       'display.max_columns', None,
                       'display.precision', 3,
                       ):
    print(data.dropna())
```
