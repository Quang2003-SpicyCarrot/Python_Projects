# Part 3: List, Set & Dictionary

- Example: list_variable = [element_1, element_2, element_3,..., element_n]

- Change one element in list: list_variable[ order_number_of_list ] = your_new_value
- Change multi-elements in list: list_variable[ start_order_number : end_order_number ] = your_new_list_values
- Delete element in list: list_variable[ order_number_of_list ] = empty_list
