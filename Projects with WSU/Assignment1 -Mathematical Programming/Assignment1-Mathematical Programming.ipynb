# Question 1: 1.1
Show that the number 142857 is cyclic, meaning if we multiply it with numbers 1 to 6, the results are the permutation of the digits of the original number 142857

```python
def Cyclic_number(a): #create a function called Cyclic_number(a)
    n = a
    for i in range(1, 7):
        n_1 = n * i
        if (sorted(str(n)) == sorted(str(n_1))): #Checking if the sorted string n matches with sorted string n_1 value.
            Cyclic = True
        else:
            Cyclic = False #if not, it will escape the for loop and return as False
            break
    return(Cyclic)
    
Cyclic_number(142857)
```

```python
#An improved version where we can add zero as initial number.
def Cyclic_number(n): #n is a string of numbers
    for i in range(1, 7):
        n_1 = int(n) * i
        if (len(str(n)) != len(str(n_1))): #since int doesn't store zero as the first value of a number, we must manually add a zero
            n_1 = str(0) + str(n_1)
        if (sorted(str(n)) == sorted(str(n_1))): #check if sorted string n matches with sorted string n_1 value.
            Cyclic = True
        else:
            Cyclic = False #if not match, break the cycle and return False, else return True
            break
    return(Cyclic)

print('Cyclic_number 142857: {}'.format(Cyclic_number(142857)))
print('Cyclic_number 052631578947368421: {}'.format(Cyclic_number('052631578947368421')))
```

# Question 2: 4.5
Plot the graphs of the functions $f(x) = 2\exp^{-x^2}$ and $g(x) = cos(sin(x) + cos(x))$ between $[-\pi, \pi]$

```python
x = np.arange(-np.pi,np.pi, 0.05) #create an array of values in-between negative pi and pi (including the negative pi and pi).
f_x = 2 * np.exp(-x**(2)) #create the function f(x)
g_x = np.cos(np.sin(x) + np.cos(x)) #create the function g(x)
plt.plot(x, f_x, x, g_x) 
plt.title('Graph of function $f(x)$ and $g(x)$')
plt.legend(['$f(x) = 2 \exp^{-x^2}$', #f(x) function is represented by the blue line
            '$g(x) = cos(sin(x) + cos(x)) $']) # g(x) function is represented by the orange line
plt.show()
```

# Question 4: 6.13
Consider two surfaces
$$ q(x, y) = cos(x^2 + y^2)\exp^{-x^2}$$
$$ w(x, y) = 3 - x^2 - y^2. $$
Plot these function

```python
fig = plt.figure()
ax = plt.axes(projection = '3d') #adding axes to the current figure and set the projection type of the axes as 3D

x = np.linspace(-3,3,50)
y = np.linspace(-3,3,50)
x, y = np.meshgrid(x,y) #create 2D arrays
q_x_y = np.cos(x**2 + y**2)*(np.exp(-x**2)) #find the values for the surface q(x,y)
w_x_y = 3 - x**2 - y**2 #find the values for the second surface w(x,y)
ax.plot_surface(x,y, q_x_y, rstride = 1, cstride = 1, cmap ='viridis') #plot the surface
ax.plot_surface(x,y, w_x_y, rstride = 1, cstride = 1, cmap ='viridis')
```

# Question 5: 8.6
Exercise 8.6: Plot the polar graph of $r = 3\cos(6\theta)$

**Answer**

Since we can find radius $r$ directly through the equation above, we can also use basic trigonometry, with sin, cos and tan, to find the adjacent length and the opposite length of the right triangle where r is the length of the hypothenuse. Then we use the side lengths of the triangle to estimate the coordination of x and y on the Cartesian plane. Therefore, we can then graph it.
Based on the concept of trigonometry, $x = r \sin(\theta)$ returns the coordinate of x and $y = r \cos(\theta)$ returns the coordinate of y on the Cartesian plane.

$$x = r\sin(\theta)$$
$$y = r\cos(\theta)$$

```python
angle = np.arange(0,360,0.005) #assign values to theta
r = 3 * np.cos(6*angle) #find r value
x = r * np.sin(angle) #find the x-coordinate
y = r * np.cos(angle) #find the y-coordinate 
plt.axis('equal')
plt.plot(x,y)
```
