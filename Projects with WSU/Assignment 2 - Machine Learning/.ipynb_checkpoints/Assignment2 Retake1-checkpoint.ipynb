# Question 2

CNN Steps: 
- Feature Learning: Input, Convolution + ReLU, Pooling,
- Feature Learning: Convolution + ReLU, Pooling, etc
- Classification: Flatten, Fully Connected, Softmax 

FORMULA TO CALCULATE THE CONVOLUTIONAL LAYER: $(W - F + 2P)/S + 1$
- W: input size (ixi)
- F: kernel size (default = 0)
- P: padding (default = 0)
- S: stride (default = 1)

FORMULA TO CALCULATE THE INVERSE CONVOLUTIONAL LAYER: $(W - 1) * S - 2P + F$

```python
class CNNAutoEncoder(nn.Module):
    def __init__(self):
        
        #N, 3, 32 * 32 = 1024
        super().__init__()
        self.encoder_layer = nn.Sequential(
            # Conv Layer block 
            nn.Conv2d(in_channels=3, out_channels=32, kernel_size=3, padding=1), #32 x 32 x 32
            nn.BatchNorm2d(32),
            nn.ReLU(inplace=True),
            
            nn.Conv2d(in_channels=32, out_channels=16, kernel_size= 3, padding=1), #16 x 32 x 32
            nn.BatchNorm2d(16),
            nn.MaxPool2d(kernel_size=2, stride=2) #16 x 16 x 16
            )
        
        self.decoder_layer = nn.Sequential(
            # Conv Layer block 
            nn.ConvTranspose2d(16, 32, 2, stride=2),
            nn.BatchNorm2d(32),
            
            nn.ReLU(inplace=True),
            nn.Conv2d(32, 3, 3, padding = 1),
            nn.ReLU(inplace = True),
            nn.BatchNorm2d(3)
            )
    
    def forward(self,x):
        encoder_val = self.encoder_layer(x)
        v = encoder_val.view(encoder_val.size(0), - 1) #in terms of dimensionality, we fix it to be 2
        x = self.decoder_layer(encoder_val)
        return x, v
```

---

```python
def train_model(optimiser, model, loss_function, trainloader, valloader, n_epochs = 10, fplotloss=True, mode=1):
    train_on_gpu = torch.cuda.is_available()
    if train_on_gpu:
        print("GPU is available! Train model on GPU.")
        model.cuda()
    
    train_losslist = []
    val_losslist = []
   
    
    # Track the best model so far (evaluated on validation set)
    val_loss_min = np.Inf 

    #enter cycle
    print("Entering training cycles.")
    for epoch in [*range(n_epochs)]:
        train_loss = 0.0
        val_loss = 0.0
        
        model.train()
        for data, target in trainloader:
            if train_on_gpu: 
                data, target = data.cuda(), target.cuda()
            
            optimiser.zero_grad()
            output = model(data)
            if mode==0:
                loss = loss_function(output, target)
            else: #AE mode training
                loss = loss_function(output[0], data)
            loss.backward()
            optimiser.step()
            train_loss += loss.item() * data.size(0)
 
        # validate the model 
        model.eval()
        for data, target in valloader:
            if train_on_gpu == True: 
                data, target = data.cuda(), target.cuda()

            output = model(data)
            if mode == 0:
                loss = loss_function(output, target)
            else: 
                loss = loss_function(output[0], data)
            val_loss += loss.item() * data.size(0)
        
        # calculate average losses
        train_loss = train_loss / len(trainloader.dataset)
        val_loss = val_loss / len(valloader.dataset)
    
        train_losslist.append(train_loss)
        val_losslist.append(val_loss)
            
        # print training/validation statistics 
        print('Epoch: {} \tTraining Loss: {:.6f} \tValidation Loss: {:.6f}'.format(epoch, train_loss, val_loss))
        
        # save model if validation loss has decreased
        if val_loss <= val_loss_min:
            print('Validation loss decreased ({:.6f} --> {:.6f})'.format(val_loss_min, val_loss))
            val_loss_min = val_loss

    
    # Plot training and validation loss if fplotloss=True
    if fplotloss == True:
        plt.plot(*range(n_epochs), train_losslist)
        plt.plot(*range(n_epochs), val_losslist)
        plt.ylim((min(train_losslist+val_losslist),max(train_losslist+val_losslist)))
        plt.xlabel("Epoch")
        plt.ylabel("Loss")
        plt.title("Performance of the model")
        plt.legend(["Training loss","Validation loss"])
        plt.show()    

    # Process is complete.
    print('Training process has finished.')
```

```python
train_loader_model = train_loader
val_loadpck_model = val_loadpck
num_epochs = 10
learning_rate = 0.1
loss_function = nn.MSELoss()
model = CNNAutoEncoder()
optimiser = torch.optim.SGD(model.parameters(), lr = learning_rate)
```

```python
train_model(optimiser = optimiser, 
            model = model,
            loss_function = loss_function,
            trainloader = train_loader_model,
            valloader = val_loadpck_model,
            n_epochs = num_epochs,
            fplotloss = True,
            mode = 1)
```

```python
import torchvision as tv
def imshow(img, ax):
    npimg = img.numpy()
    ax.imshow(np.transpose(npimg, (1, 2, 0)))
    ax.axis('off')
```

```python
n = 5
torch.manual_seed(3)
dataiter = iter(train_loader)
imgs, lbls = next(dataiter)
output, v = model(imgs)


plt.figure(figsize=(20,20))
fig, axes = plt.subplots(2,n)
for i in range(n):
    imshow(tv.utils.make_grid(imgs[i]),axes[0,i])
    axes[0,i].set_title(classes[int(lbls[i])])

    im = output[i].detach()
    imshow(tv.utils.make_grid(im),axes[1,i])
    if i==int(n/2):
        axes[1,i].set_title('Reconstructed images')

fig.tight_layout()
plt.show()
```

```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import pandas as pd

#train set
train_features = []
train_features_labels = []

torch.manual_seed(3)
for i in range(len(train_loader)):
    dataiter = iter(train_loader)
    imgs, lbls = next(dataiter)
    output, v = model(imgs)
    train_features = train_features + v.detach().numpy().tolist()
    train_features_labels = train_features_labels + lbls.detach().numpy().tolist()


train_df = pd.DataFrame(data = train_features)
train_df_labels = pd.DataFrame(data = train_features)
train_df_labels['labels'] = train_features_labels


#test set
test_features = []
test_features_labels = []

torch.manual_seed(3)
for i in range(1):
    dataiter = iter(val_loadpck)
    imgs, lbls = next(dataiter)
    output, v = model(imgs)
    test_features = test_features + v.detach().numpy().tolist()
    test_features_labels = test_features_labels + lbls.detach().numpy().tolist()


test_df = pd.DataFrame(data = test_features)
test_df_labels = pd.DataFrame(data = test_features)
test_df_labels['labels'] = test_features_labels
```

```python
from sklearn.manifold import TSNE

#for the training set
X_embedded = TSNE(n_components = 3, learning_rate = 0.1, init = 'random', perplexity = 0.5).fit_transform(train_df)
X_embedded = pd.DataFrame(data = X_embedded)

X_sym = {2: '+', 1: '>' , 8: 'X'}
X_col = {2: 'black', 1: 'red', 8:'green'}
X_embedded['labels'] = train_features_labels
X_embedded['color'] = X_embedded['labels'].replace(X_col)
X_embedded['symbols'] = X_embedded['labels'].replace(X_sym)
X_embedded


#for the testing set
X_embedded_test = TSNE(n_components = 2, learning_rate = 0.1, init = 'random', perplexity = 0.5).fit_transform(test_df)
X_embedded_test = pd.DataFrame(data = X_embedded_test)

X_embedded_test['labels'] = test_features_labels
X_embedded_test['color'] = X_embedded_test['labels'].replace(X_col)
X_embedded_test['symbols'] = X_embedded_test['labels'].replace(X_sym)
X_embedded_test

X_embedded
```
